```python
import sys
import pandas
import numpy 
import os.path
import math
import itertools
sys.path.append("..")    

# New data frame
data = pandas.read_csv('geneva/schedule-real-time.csv', header=0)
```

```python
trimmed_data = data.drop(['passengerCountDoor1Up', 'passengerCountDoor2Up', 'passengerCountDoor3Up', 'passengerCountDoor4Up', 'passengerCountDoor5Up', 
'passengerCountDoor6Up', 'passengerCountDoor1Down', 'passengerCountDoor2Down', 'passengerCountDoor3Down', 
'passengerCountDoor4Down', 'passengerCountDoor5Down', 'passengerCountDoor6Down'], axis=1)

trimmed_data = trimmed_data.drop(['isRedirection','isDelocalisation','isRelocalisation'], axis=1)

trimmed_data = trimmed_data.drop(['patternId'], axis=1)
```

```python
# Now that we have the trimmed data, let's sort it per some rules.
sorted_data = trimmed_data.sort(['date','routeCode','tripId','tripDirection','stopSequenceNr'],axis=0)
```

```python
# Now let's load the stops mappping file.
stops_mapping = pandas.read_csv('geneva/gva-stops-mapping.csv', header=0)
stops_mapping.tail()
```

```python
merged_data = sorted_data.merge(stops_mapping, on='stopCode')
merged_data = merged_data.sort(['date','routeCode','tripId','tripDirection','stopSequenceNr'],axis=0)
merged_data = merged_data.drop(['serviceVehicle','passengerCountTripUp','passengerCountTripDown','passengerCountStopUp','passengerCountStopDown'], axis=1)
```

```python
# Now we want to do some calculations on the data that will be useful for frustration calculation.
# We need speed from previous stop (at each stop), number of people that got on, and delay at each stop.

# We have load. 
# Let's try getting the delay at that particular stop.

# define a function for calculating delay. 
from datetime import datetime 
import time
def delay(scheduledTime, actualTime):
    if scheduledTime.lower() == "nan" or actualTime.lower() == "nan":
        return "nan"
    
    st = datetime.strptime(scheduledTime, "%Y-%m-%d %H:%M:%S")
    at = datetime.strptime(actualTime, "%Y-%m-%d %H:%M:%S")
    
    # Find the difference between the 2 (actual time - scheduled time).
    return time.mktime(at.timetuple()) - time.mktime(st.timetuple())                                           
```

```python
# Create an empty data frame.
df = pandas.DataFrame(index=merged_data.index, columns=['delay'])
df = df.fillna(0)

# iterate through merged data, and calculate the delay
for tup in merged_data.itertuples():
    idx = tup[0]
    sched = str(tup[10])
    actual = str(tup[11])
    time_delay = delay(sched,actual)
    if idx != 0:
        df.delay[idx] = time_delay
    else:
        df.delay[idx] = 0
```

```python
# How the hell do I end up with delay in the merged data.
temp = pandas.concat([merged_data, df], axis=1)
```

```python
import math

def distance_on_unit_sphere(lat1, long1, lat2, long2):

    # Convert latitude and longitude to 
    # spherical coordinates in radians.
    degrees_to_radians = math.pi/180.0
        
    # phi = 90 - latitude
    phi1 = (90.0 - lat1)*degrees_to_radians
    phi2 = (90.0 - lat2)*degrees_to_radians
        
    # theta = longitude
    theta1 = long1*degrees_to_radians
    theta2 = long2*degrees_to_radians
        
    # Compute spherical distance from spherical coordinates.
        
    # For two locations in spherical coordinates 
    # (1, theta, phi) and (1, theta, phi)
    # cosine( arc length ) = 
    #    sin phi sin phi' cos(theta-theta') + cos phi cos phi'
    # distance = rho * arc length
    
    cos = (math.sin(phi1)*math.sin(phi2)*math.cos(theta1 - theta2) + 
           math.cos(phi1)*math.cos(phi2))
    arc = math.acos( cos )

    # Remember to multiply arc by the radius of the earth 
    # in your favorite set of units to get length.
    return arc
```

```python
def distance(lat1, lon1, lat2, lon2):
    try:
        # return distance in miles, and kilometers
            unit_distance = distance_on_unit_sphere(lat1, lon1, lat2, lon2) 

        # need to multiply by 3690 for distance in miles, 6373 for distance in km.
        # return an array with [dist in miles, dist in km]
            return unit_distance*3960
    except ValueError:
        print str(lat1)+" "+str(lon1)+" "+str(lat2)+" "+str(lon2)
```

```python
# Create an empty data frame.
distf = pandas.DataFrame(index=temp.index, columns=['dist'])
distf = distf.fillna(0)
i = 0
# iterate through merged data, and calculate the distance from the lat/lon in the previous row
prevTup = ()
for tup in merged_data.itertuples():
    if len(prevTup) == 0:
        prevTup = tup
    
    idx = tup[0]
    currLat = float(tup[13])
    currLon = float(tup[14])
    
    prevLat = float(prevTup[13])
    prevLon = float(prevTup[14])
    
    curr_distance = str(distance(currLat,currLon,prevLat,prevLon))
    if i < 41:
        print curr_distance + " "+str(currLat) +" "+str(currLon) + " "+str(prevLon) +" "+str(prevLat)
    
    if idx != 0:
        distf.dist[idx] = curr_distance
    else:
        distf.dist[idx] = 0
        
    prevTup = tup
    i = i+1
```

```python
def timeFromPrev(currTime, prevTime):
    if currTime.lower() == "nan" or prevTime.lower() == "nan":
        return "nan"
    
    ct = datetime.strptime(currTime, "%Y-%m-%d %H:%M:%S")
    pt = datetime.strptime(prevTime, "%Y-%m-%d %H:%M:%S")
    
    # Find the difference between the 2 (actual time - scheduled time).
    return time.mktime(ct.timetuple()) - time.mktime(pt.timetuple()) 
```

```python
# Now we want to do calculate time between stops. What what.
# Create an empty data frame.
timef = pandas.DataFrame(index=temp.index, columns=['timeFromPrevStop'])
timef = timef.fillna(0)
i = 0
# iterate through merged data, and calculate the distance from the lat/lon in the previous row
prevTup = ()
for tup in merged_data.itertuples():
    if len(prevTup) == 0:
        prevTup = tup
    
    idx = tup[0]
    prevStopArrivalTime = prevTup[11]
    currStopArrivalTime = tup[11]
    
    timeFromPrevStop = timeFromPrev(currStopArrivalTime,prevStopArrivalTime)
    if idx != 0:
        timef.timeFromPrevStop[idx] = timeFromPrevStop
    else:
        timef.timeFromPrevStop[idx] = 0
        
    prevTup = tup
    i = i+1
    
```

```python
def speed(row):
    try:
        if (int(row[17])== 0 or int(row[17])== None or str(row[17]).lower() =="nan" or row[17]=="None" or row[16] == "nan" or row[16] == None or row[16] == "None"):
            return "nan"
    
        return float(row[16])*1609.34/float(row[17])
    except ValueError:
        print str(row[17]) + " " + str(row[16])
```

```python
# THE SPEED IS GOING TO BE IN METRES PER SECOND. THIS WAS TO PREVENT UNDERFLOW ERRORS.

# Now we want to do calculate speed between stops. What what.
# Create an empty data frame.
speedf = pandas.DataFrame(index=temp.index, columns=['speedFromPrevStop'])
speedf = speedf.fillna(0)

# iterate through merged data, and calculate the speed for each row.
for tup in temp.itertuples():
    idx = tup[0]
    speedFromPrevStop = speed(tup)
    if idx != 0:
        speedf.speedFromPrevStop[idx] = speedFromPrevStop
    else:
        speedf.speedFromPrevStop[idx] = 0
```

```python
final_gva = pandas.concat([temp, speedf], axis=1)

final_gva.tail(1)
```

```python
# Now we calculate the frustration, but that comes in due course. (and limit it to Monday and such) Might need to sort by time and all that.

# Calculate the capacity based on the average passengerLoadStop for each route.
routeGroup = final_gva.groupby(final_gva.routeCode)
```

```python
routesToFrame = dict()
i = 0
tempgroup = pandas.DataFrame()
for name, group in routeGroup:
    #print name
    if i == 0:
        tempgroup = group
    routesToFrame[name] = group
    i = 1

tempgroup.tail()
```

```python
routesToAvgLoad = dict()
for key in routesToFrame.keys():
    curr_df = routesToFrame[key]
    mean = curr_df['passengerLoadStop'].mean()
    routesToAvgLoad[key] = mean
```

```python
# Create an empty data frame.
capacityf = pandas.DataFrame(index=final_gva.index, columns=['capacity'])
capacityf = capacityf.fillna(0)

# iterate through merged data, and calculate the speed for each row.
for tup in final_gva.itertuples():
    idx = tup[0]
    # get average load for this route.
    currRouteCode = tup[2]
    if str(currRouteCode) != "nan":
        avgload = routesToAvgLoad[currRouteCode]
        currLoad = int(tup[12])
        currCapacity = round((float(currLoad)/float(avgload)),2)
        if idx != 0:
            capacityf.capacity[idx] = currCapacity
        else:
            capacityf.capacity[idx] = 0
    else:
        capacityf.capacity[idx] = -1
```

```python
final_gva = pandas.concat([final_gva, capacityf], axis=1)

final_gva.tail()
```
