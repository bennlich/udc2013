```python
# Now we want to do some calculations on the data that will be useful for frustration calculation.

# Let's try getting the delay at that particular stop.

# define a function for calculating delay. 
from datetime import datetime 
import time
def delay(scheduledTime, actualTime):
    if scheduledTime.lower() == "nan" or actualTime.lower() == "nan" or scheduledTime == "-1" or actualTime == "-1":
        return "nan"
    
    return int(actualTime) - int(scheduledTime)                                           
```

```python
# Create an empty data frame.
df = pandas.DataFrame(index=merged_sched_data.index, columns=['delay'])
df = df.fillna(0)

# iterate through merged data, and calculate the delay
for tup in merged_sched_data.itertuples():
    idx = tup[0]
    sched = str(tup[13])
    actual = str(tup[15])
    time_delay = delay(sched,actual)
    if idx != 0:
        df.delay[idx] = time_delay
    else:
        df.delay[idx] = 0
```

```python
import math

def distance_on_unit_sphere(lat1, long1, lat2, long2):

    # Convert latitude and longitude to 
    # spherical coordinates in radians.
    degrees_to_radians = math.pi/180.0
        
    # phi = 90 - latitude
    phi1 = (90.0 - lat1)*degrees_to_radians
    phi2 = (90.0 - lat2)*degrees_to_radians
        
    # theta = longitude
    theta1 = long1*degrees_to_radians
    theta2 = long2*degrees_to_radians
        
    # Compute spherical distance from spherical coordinates.
        
    # For two locations in spherical coordinates 
    # (1, theta, phi) and (1, theta, phi)
    # cosine( arc length ) = 
    #    sin phi sin phi' cos(theta-theta') + cos phi cos phi'
    # distance = rho * arc length
    
    cos = (math.sin(phi1)*math.sin(phi2)*math.cos(theta1 - theta2) + 
           math.cos(phi1)*math.cos(phi2))
    arc = math.acos( cos )

    # Remember to multiply arc by the radius of the earth 
    # in your favorite set of units to get length.
    return arc
```

```python
def distance(lat1, lon1, lat2, lon2):
    try:
        # return distance in miles, and kilometers
            unit_distance = distance_on_unit_sphere(lat1, lon1, lat2, lon2) 

        # need to multiply by 3690 for distance in miles, 6373 for distance in km.
        # return an array with [dist in miles, dist in km]
            return unit_distance*3960
    except ValueError:
        print str(lat1)+" "+str(lon1)+" "+str(lat2)+" "+str(lon2)
```

```python
def timeFromPrev(currTime, prevTime):
    return int(currTime) - int(prevTime)
```

```python
def speed(row):
    try:
        if (int(row[22])== 0 or int(row[22])== None or str(row[22]).lower() =="nan" or row[22]=="None" or row[21] == "nan" or row[21] == None or row[21] == "None"):
            return "nan"
    
        return float(row[21])*1609.34/float(row[22])
    except ValueError:
        print str(row[22]) + " " + str(row[21])
```

```python
# THE SPEED IS GOING TO BE IN METRES PER SECOND. THIS WAS TO PREVENT UNDERFLOW ERRORS.

# Now we want to do calculate speed between stops. What what.
# Create an empty data frame.
speedf = pandas.DataFrame(index=temp.index, columns=['speedFromPrevStop'])
speedf = speedf.fillna(0)

# iterate through merged data, and calculate the speed for each row.
for tup in temp.itertuples():
    idx = tup[0]
    speedFromPrevStop = speed(tup)
    if idx != 0:
        speedf.speedFromPrevStop[idx] = speedFromPrevStop
    else:
        speedf.speedFromPrevStop[idx] = 0
```
